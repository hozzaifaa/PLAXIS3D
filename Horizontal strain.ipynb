```python
from plxscripting.easy import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#close all windows then open the new windows
s_o, g_o = new_server('localhost', g_i.view(phase_1), password=s_i.connection._password)

# Initialize the figure and axes
fig, ax = plt.subplots(figsize=(15, 5))

# Loop through phases 1 to 4
for phase_num in range(1, 4):
    phase_o = g_o.Phases[phase_num]
    plot_o = g_o.crosssectionplot(g_o.Plots[0], (xmin, ymax/2, 0), (xmax, ymax/2, 0), (xmin, ymax/2, 1))
    
    # Extract cross-section results for X, Z, and Uz
    r_X = np.array(g_o.getcrosssectionresults(plot_o, phase_o, g_o.ResultTypes.Soil.X, True))
    r_Z = np.array(g_o.getcrosssectionresults(plot_o, phase_o, g_o.ResultTypes.Soil.Z, True))
    r_Uz = np.array(g_o.getcrosssectionresults(plot_o, phase_o, g_o.ResultTypes.Soil.Uz, True)) * 1000
    
    # Create a DataFrame for the current phase
    data = pd.DataFrame({
        "X": r_X,
        "Z": r_Z,
        "Uz": r_Uz
    })
    data = data.loc[data.Z == 0]
    data = data.sort_values(by=['X', 'Z']).reset_index(drop=True).round({'X': 2, 'Z': 0}).drop_duplicates(subset=['X'], keep='first', ignore_index=True)
    
    # Calculate settlement volume and volume loss for the current phase
    data['ΔUz'] = -(data.iloc[:, 2] + data.iloc[:, 2].shift(1)) / 2
    data['ΔX'] = (data.iloc[:, 0] - data.iloc[:, 0].shift(1))
    data['Vs'] = (data['ΔX'] * data['ΔUz'])
    V_T = np.pi * radius**2
    V_S = data['Vs'].sum() / 1000
    V_L = V_S / V_T
    
    # Plot the data for the current phase
    data.plot(x='X', y='Uz', linestyle='-', linewidth=1, ax=ax, label=f"FEM - Phase {phase_num} - max: {(abs(data['Uz'].min())):.2f} mm, [VL: {V_L:.2%}]")

# Plot the Gaussian curve (based on the last phase's Uz_max)
Uz_max = -50.17
K = 0.54
H = 23
i = K * H
x_values = np.arange(xmin, xmax, 0.1)
Uz = Uz_max * np.exp(-(x_values**2) / (2 * i**2))
ax.plot(x_values, Uz, linestyle='-', color='k', linewidth=1, label=f'Gaussian curve (k={K})')

# Plot the Yiu et al. 2017 curve
data_Yiu = pd.read_csv(r"C:\Users\hozza\Desktop\My project\Yiu et al. 2017.csv")
data_Yiu.plot(x='X', y='Uz', linestyle='-.', color='k', linewidth=1, ax=ax, label=f"Yiu et al. 2017 - max: {(abs(data_Yiu['Uz'].min())):.2f} mm,[VL: 1.65%]")

# Add titles, labels, grid, and legend
ax.set_title(f"Greenfield simulation with phases 1 to 4, Gaussian curve, and Yiu et al. 2017")
ax.set_xlabel('Horizontal distance [m]')
ax.set_ylabel('Vertical displacement [mm]')
ax.grid(True)
ax.legend()

# Set x-axis limits
plt.xlim(-60, 60)

# Show the plot
plt.show()
```
